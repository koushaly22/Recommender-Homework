```python
import pandas as pd
import numpy as np
```

```python
#Class generated data - downloaded 11/12/2016 to Movie_Reviews.csv
M = pd.read_csv("Movie_Reviews.csv", index_col=0)
```

```python
M
```

```python
def pearson(s1, s2):
    """Take two pd.Series objects and return a pearson correlation."""
    s1_c = s1 - s1.mean()
    s2_c = s2 - s2.mean()
    return np.sum(s1_c * s2_c) / np.sqrt(np.sum(s1_c ** 2) * np.sum(s2_c ** 2))
```

```python
def get_recs(movie_name, M, num):

    import numpy as np
    reviews = []
    for title in M.columns:
        if title == movie_name:
            continue
        cor = pearson(M[movie_name], M[title])
        if np.isnan(cor):
            continue
        else:
            reviews.append((title, cor))
    
    reviews.sort(key=lambda tup: tup[1], reverse=True)
    return reviews[:num]
```

<h3>Question 1: What movie is most Similar to 'The fault in our stars' (60 pts)</h3>

```python
recs = get_recs('The Fault in Our Stars', M, 10)
```
